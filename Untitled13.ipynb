```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([0,0,5,5])
y=np.array([0,35,35,0])
mycolor=['red','green','yellow','lime']
size=[10,60,120,80]

plt.scatter(x,y,color=mycolor,s=size)
plt.show()
```
